```python
import time
from vllm import LLM, SamplingParams

path_to_model = "/media/msst/ssd_storage1/ml/llm/pretrained_models/Llama2-7B"

llm = LLM(
    model=path_to_model,
    dtype="float16",
    max_model_len=1024
)

params = SamplingParams(
    max_tokens=512,
    temperature=0.0
)

prompt = "Объясни, что такое LLM."

# warm-up
llm.generate([prompt], params)

start = time.time()
out = llm.generate([prompt], params)
elapsed = time.time() - start

tokens = len(out[0].outputs[0].token_ids)
print(f"Speed: {tokens / elapsed:.2f} token/s")
```
